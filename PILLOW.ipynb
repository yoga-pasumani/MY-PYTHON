```python
import cv2
from google.colab.patches import cv2_imshow

# Load image
img = cv2.imread("/content/ChatGPT Image May 19, 2026, 05_48_21 PM.png")

# Convert to grayscale (important for detection)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Load pre-trained face detector
face_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
)

# Detect faces
faces = face_cascade.detectMultiScale(
    gray,
    scaleFactor=1.1,
    minNeighbors=5
)

# Draw rectangles
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)

# Show output
cv2_imshow(img)
```

cv2.imread() → loads image
cvtColor() → converts to black & white (easier detection)
CascadeClassifier() → pretrained face detector brain
detectMultiScale() → finds faces
rectangle() → draws box
cv2_imshow() → shows result in Colab

```python
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas

# Create PDF file
file_name = "resume.pdf"
c = canvas.Canvas(file_name, pagesize=letter)

# Start writing content
c.setFont("Helvetica-Bold", 18)
c.drawString(200, 750, "RESUME")

# User details (you can take input later)
name = "Your Name"
phone = "1234567890"
email = "yourmail@gmail.com"
skills = "Python, ML, SQL"
education = "MSc Data Analytics"
experience = "Fresher"

c.setFont("Helvetica", 12)

c.drawString(100, 700, f"Name: {name}")
c.drawString(100, 680, f"Phone: {phone}")
c.drawString(100, 660, f"Email: {email}")
c.drawString(100, 640, f"Skills: {skills}")
c.drawString(100, 620, f"Education: {education}")
c.drawString(100, 600, f"Experience: {experience}")

# Save PDF
c.save()

print("Resume created successfully!")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

# Dataset
data = {
    "text": [
        "Government announces new education policy",
        "Aliens landed in Chennai yesterday",
        "India wins cricket match",
        "Doctors confirm miracle cure for all diseases"
    ],
    "label": [1, 0, 1, 0]
}

df = pd.DataFrame(data)

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    df["text"], df["label"], test_size=0.2
)

# Convert text → numbers
vectorizer = TfidfVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

# Model
model = LogisticRegression()
model.fit(X_train_vec, y_train)

# Test prediction
sample = ["Scientists discover water on Mars"]
sample_vec = vectorizer.transform(sample)

prediction = model.predict(sample_vec)

print("Prediction:", "REAL 🟢" if prediction[0] == 1 else "FAKE 🔴")
```
